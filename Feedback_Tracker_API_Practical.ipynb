# Feedback Tracker API Practical Assignment

## Project: Feedback Tracker API

## 1. Objective

Build a backend API where users can register/login and perform secure CRUD operations on feedback entries.

## 2. Tools and Technologies

- Node.js
- Express.js
- MongoDB
- Mongoose
- JWT (JSON Web Token)
- BcryptJS
- Helmet and CORS middleware
- Postman
- Nodemon

## 3. System Architecture Diagram (Text)

Client (Postman/Web App) -> Feedback API Server (Express) -> MongoDB Database

### Explanation

- The client sends API requests.
- The Express server validates, authorizes, and processes data.
- MongoDB stores users and feedback records.
- Middleware handles logging, security, and errors.

## 4. Project Folder Structure

```text
feedbackTracker/
  src/
    config/
      db.js
    controllers/
      authController.js
      feedbackController.js
    middleware/
      authMiddleware.js
      errorMiddleware.js
      loggerMiddleware.js
    models/
      User.js
      Feedback.js
    routes/
      authRoutes.js
      feedbackRoutes.js
    app.js
    server.js
```

This organization separates business logic, routing, and data models.

## 5. Database Schema

### User Schema

- name: User full name
- email: Unique email for login
- password: Encrypted password
- role: user/admin (if used)

### Feedback Schema

- user: Reference to the user who submitted feedback
- title: Short feedback title
- message: Detailed feedback text
- rating: Numeric rating (for example 1 to 5)
- createdAt/updatedAt: Timestamp fields

References help connect feedback data with user details.

## 6. API Endpoints Table

| Method | Endpoint | Purpose |
|---|---|---|
| POST | /api/auth/register | Register user |
| POST | /api/auth/login | Login and get token |
| POST | /api/feedback | Create feedback (auth) |
| GET | /api/feedback | List feedback (auth) |
| GET | /api/feedback/:id | Get one feedback (auth) |
| PUT | /api/feedback/:id | Update feedback (auth) |
| DELETE | /api/feedback/:id | Delete feedback (auth) |
| GET | /api/feedback/filter?rating=4 | Filter by rating (auth) |
| GET | /api/feedback/stats | Feedback statistics (auth) |

## 7. Sample Request and Response

### Sample Request: Create Feedback

```http
POST /api/feedback
Authorization: Bearer <feedback_token>
Content-Type: application/json

{
  "title": "Support Team",
  "message": "Quick response and clear guidance",
  "rating": 5
}
```

### Sample Response

```json
{
  "message": "Feedback created successfully",
  "feedback": {
    "_id": "6640b1231234abcd5678ef90",
    "title": "Support Team",
    "message": "Quick response and clear guidance",
    "rating": 5,
    "user": "6640a9991234abcd5678ef22"
  }
}
```
